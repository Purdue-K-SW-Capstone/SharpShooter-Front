```python
import soundfile as sf
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
###############Calculate the length of data set###############
seconds = []
sec_sum=0
count=0
for _ in glob.glob('./youtube-audio(wav)/*.wav'):
    f = sf.SoundFile(_)
    sec = f.frames / f.samplerate
    seconds.append(round(sec,2))
    sec_sum+=sec
    count+=1
    
for _ in glob.glob('./youtube-auido(wav)_yb/*.wav'):
    f = sf.SoundFile(_)
    sec = f.frames / f.samplerate
    seconds.append(round(sec,2))
    sec_sum+=sec
    count+=1

    
for _ in glob.glob('./coyote_dataset_montana/*.wav'):
    f = sf.SoundFile(_)
    sec = f.frames / f.samplerate
    seconds.append(round(sec,2))
    sec_sum+=sec
    count+=1
        
for _ in glob.glob('./coyote_dataset/*.wav'):
    f = sf.SoundFile(_)
    sec = f.frames / f.samplerate
    seconds.append(round(sec,2))
    sec_sum+=sec
    count+=1
print(sec_sum)
print(sec_sum/3)    
print(sec_sum/5)
```

```python
#######################split data######################
# First load the file
import librosa
import os
import soundfile as sf

dir_path = "/Users/bbo/Downloads/dogs/"

for (root, directories, files) in os.walk(dir_path):
    for file in files:
        if '.wav' in file:
            file_path = os.path.join(root, file)
            file_name =os.path.basename(file_path)
            #file_path = '/Users/bbo/Downloads/coyote_dataset/coyote_calls_day.wav'
            audio, sr = librosa.load(file_path)
            # Get number of samples for 5 seconds; 
            buffer = 5 * sr

            samples_total = len(audio)
            samples_wrote = 0
            counter = 1

            while samples_wrote < samples_total:
                #check if the buffer is not exceeding total samples 
                if buffer > (samples_total - samples_wrote):
                    buffer = samples_total - samples_wrote

                block = audio[samples_wrote : (samples_wrote + buffer)]
                #change directory and save the file
                os.chdir('/Users/bbo/Downloads/save2')
                out_filename = "dog_" + str(counter) + ".wav" 

                sf.write(out_filename, block, sr)
                counter += 1
                samples_wrote += buffer
```

```python
#######################waveform code ver 1#####################
import librosa
import librosa.display
import matplotlib.pyplot as plt


FIG_SIZE = (30, 20)
file = "split_2_coyote.wav"

# load audio file with Librosa
signal, sample_rate = librosa.load(file, sr=48000)
print('signal shape : ', signal.shape)

# WAVEFORM
plt.figure(figsize=FIG_SIZE)
librosa.display.waveshow(signal, sample_rate, alpha=0.4)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```
